## Polynomial Curve Fitting

Begin with a simple regression problem. Suppose we have a real-valued input variable $x$ and we want to use this observation to predict the value of a real-valued target variable $t$. Let's use some generated data from the function $\sin(2\pi x)$ with random noise included in the target values. Suppose we are given a training set of $N$ observations of $x$, written as $\mathbf{x}\equiv(x_1,...,x_N)^\mathrm{T}$ with corresponding observations of the values of $t$, denoted $\mathbf{t}\equiv(t_1,...,t_N)^\mathrm{T}$.

```python
import numpy as np
import matplotlib.pyplot as plt

N_SAMPLES = 10

# Generate x and y values
x = np.array([np.random.uniform(0., 1.) for i in range(N_SAMPLES)])
y = np.sin(2*np.pi*x)

# Add random (Gaussian) noise to signal
NOISE_LEVEL = 0.1
y = [yy + np.random.normal(0., NOISE_LEVEL) for yy in y]

# Plot the data
plt.scatter(x, y)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y(x)$', fontsize=12)
plt.show()
```

We want to use this training set to make predictions of the value of $\widehat{t}$ of the target variable for some new value $\widehat{x}$ of the target variable for some new value $\widehat{x}$ of the input variable - this involves trying to discover the underlying functin $\sin(2\pi x)$.

Consider a simple approach based on curve fitting, and use a polynomial function of the form
$$
y(x, \mathbf{w}) = w_0 + w_{1}x + w_{2}x^2 + ... + w_{M}x^M = \sum^{M}_{j=0}w_jx^j
$$
where $M$ is the order of the polynomial. The polynomial coefficients $w_0, ..., w_m$ are collectively denoted by the vector $\mathbf{w}$. Although the polynomial function $y(x, \mathbf{w})$ is a nonlinear function of $x$, it is a linear function of the coefficients $\mathbf{w}$. Functions that are linear in the unknown parameters are called *linear models*.

The values of the coefficients will be determined by fitting the polynomial to the training data. This can be done my minimising an *error function* that measures the misfit between $y(x, \mathbf{w})$, for any given value of $\mathbf{w}$, and the training set of data points. A simple choice of error function is given by the sum of squares of the errors between the predictions $y(x_n, \mathbf{w})$ for each data point $x_n$ and the corresponding target values $t_n$, so that we minimise
$$
E(\mathbf{w}) = \frac{1}{2}\sum^{N}_{n=1}\lbrace y(x_n, \mathbf{w}) - t_n) \rbrace^2
$$
where the factor $1/2$ is included for later convenience. Note that the errpr function is a nonnegative quantity that would be zero if, and only if, $y(x, \mathbf{w})$ were to pass through exactly each training data point.

We can solve the curve fitting problem by choosing the value of $\mathbf{w}$ for which $E(\mathbf{w})$ is as small as possible. The error function is a quadratic function of the coefficients $\mathbf{w}$, its derivatives with respect to the coefficients will be linear in the elements of $\mathbf{w}$, so the minimisation of the error function has a unique solution denoted by $\mathbf{w}^\star$. The resulting polynomial is given by the function $y(x,\mathbf{w}^\star)$.

Here, we will use the `numpy` polynomial fitting method - we go through a more formal work through later. We try polynomial fits for $M=0,1,3,9$. The problem of choosing the order $M$ of the polynomial is an example of an important concept called *model comparison* or *model selection*.

```python
degrees = [0, 1, 3, 9]

fig, axs = plt.subplots(1, 4, figsize=(18, 4))

x_pred = np.arange(0.,1.0,0.01)
for i, ax in enumerate(axs):
    
    poly = np.polyfit(x, y, degrees[i])
    y_pred = np.poly1d(poly)(x_pred)
    ax.plot(x_pred, y_pred, 'r')
    ax.scatter(x, y)
    ax.set_title('$M={}$'.format(degrees[i]))
    ax.set_xlabel('$x$', fontsize=12)
    ax.set_ylabel('$y(x)$', fontsize=12)
    ax.set_ylim(-1.2,1.2)

plt.tight_layout()
plt.show()
```

The constant ($M=0$) and first order ($M=1$) polynomials give poor fits to the data and consequently poor representations of the function $\sin(2\pi x)$. The third order ($M=3$) polynomial seems to give the best fit to the function $\sin(2\pi x)$. When we use a higher order polynomial ($M=9$), we obtain an excellent fit to the training data. The polynomial passes through exactly each data point and $E(\mathbf{w}^\star)=0$. However, the fitted curve oscillates wildly and gives a poor representation of the function $\sin(2\pi x)$ - this behaviour is known as *overfitting*.

The goal is to achieve good generalisation by making accurate predictions for new data. We can obtain some quantitative insight into the dependence of generalisation performance on $M$ by considering a seperate test set comprising 100 data points generated as the set of of training set points, with new choices for the random noise values.

```python
N_SAMPLES = 100

# Generate x and y values
x100 = np.array([np.random.uniform(0., 1.) for i in range(N_SAMPLES)])
y100 = np.sin(2*np.pi*x100)

# Add random (Gaussian) noise to signal
NOISE_LEVEL = 0.2
y100 = [yy + np.random.normal(0., NOISE_LEVEL) for yy in y100]

# Plot the data
plt.scatter(x100, y100)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$y(x)$', fontsize=12)
plt.show()
```

For each choice of $M$, we can evaluate the residual value of $E(\mathbf{w}^\star)$ for the training data, and also evaluate $E(\mathbf{w}^\star)$ for the test data set. It can sometimes be more convenient to use the root-mean-square (RMS) error defined by
$$
E_{\mathrm{RMS}}=\sqrt{2E(\mathbf{w}^{\star})/N}
$$
where the division by $N$ allows us to compare different sizes of data sets on an equal footing, and the square root ensures that $E_{\mathrm{RMS}}$ is measured on the same scale as the target variable $t$.

```python
degrees = np.arange(0, 10)

#fig, axs = plt.subplots(1, 4, figsize=(18, 4))

def error_function(predicted, true):
    
    return 0.5*(np.sum( (predicted - true)**2 ))

x_pred = np.arange(0., 1.0, 0.01)

rms_list = []
weights = []

for m in degrees:
    
    n_train = 10
    n_test = 100
    
    poly = np.polyfit(x, y, m)
    weights.append(poly) # Get list of coefficients
    
    y_pred_train = np.poly1d(poly)(x)
    y_pred_test = np.poly1d(poly)(x100)
    
    train_err = error_function(y_pred_train, y)
    train_rms = np.sqrt((2*train_err)/n_train)
    
    test_err = error_function(y_pred_test, y100)
    test_rms = np.sqrt((2*test_err)/n_test)
    
    rms_list.append([train_rms, test_rms])
    
plt.plot(degrees, np.array(rms_list)[:,0], marker='o', label='Training')
plt.plot(degrees, np.array(rms_list)[:,1], 'ro', linestyle='-', label='Test')
plt.xlabel('$M$', fontsize=12)
plt.ylabel('$E_{\mathrm{RMS}}$', fontsize=12)
plt.ylim(0, 2.)

plt.legend(loc='upper left')
plt.show()
```

The test set error is a measure of how well we are doing in predicting the values of $t$ for new data observations $x$. Small values of $M$ give relatively large values for test set error, and can be attributed to the fact that the corresponding polynomials are inflexible and incapable of capturing the oscillations in the function $\sin(2\pi x)$. Values of $M$ in the range $3\leq M \leq7$ give small values for the test set error, and also give reasonable representations for the generating function $\sin(2\pi x)$. For $M=9$ the training set error goes to zero, however the test set error has become very large.

We can examine the magnitude of coefficients $\mathbf{w}^\star$ obtained from polynomials of various order.

```python
for coeff in weights:
    string = ['w{} : {:.2f}'.format(i, c) for i ,c in enumerate(coeff)]
    print('M={}'.format(len(coeff)-1), string)
```

```python
def generate_xy(n_samples, noise_level=0.1):
    
    # Generate x and y values
    x = np.array([np.random.uniform(0., 1.) for i in range(n_samples)])
    y = np.sin(2*np.pi*x)

    # Add random (Gaussian) noise to signal
    y = [yy + np.random.normal(0., noise_level) for yy in y]
    
    return x,y

x15, y15 = generate_xy(15, noise_level=0.2)
x100, y100 = generate_xy(100, noise_level=0.2)

m9_15 = np.polyfit(x15, y15, 9)
m9_100 = np.polyfit(x100, y100, 9)

x_pred = np.arange(0, 1, 0.01)
y15_pred = np.poly1d(m9_15)(x_pred)
y100_pred = np.poly1d(m9_100)(x_pred)

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].scatter(x15, y15)
axs[0].plot(x_pred, y15_pred, 'r')
axs[0].set_ylim(-1.6, 1.6)
axs[0].set_title('N=15')
axs[0].set_xlabel('$x$', fontsize=12)
axs[0].set_ylabel('$y(x)$', fontsize=12)

axs[1].scatter(x100, y100)
axs[1].plot(x_pred, y100_pred, 'r')
axs[1].set_ylim(-1.6, 1.6)
axs[1].set_title('N=100')
axs[1].set_xlabel('$x$', fontsize=12)
axs[1].set_ylabel('$y(x)$', fontsize=12)

plt.tight_layout()
plt.show()
```

As $M$ increases, the magnitude of coefficients typically gets larger. For the $M=9$ polynomial, the coefficients have become finely tuned to the data by developing large positive and negative values so that the corresponding polynomial function matches each of the data points exactly, but between data points the function exhibits large osciilations. What is happening is that the more flexible polynomials with larger values of $M$ are becoming increasingly tuned to the random noise on the target values. 

For a given model complexity, the over-fitting problem becomes les ssever as the size of the data set increases. In other words, the larger the data set, the more complex the model that we can afford to fit to the data. A rough heuristic that is sometimes advocated is that the number of data points should be no less that some multiple (5 or 10) of the number of adaptive parameters in the model. 

It would seem more reasonable to choose the complexity of the model according to the complexity of the problem being solved. The least squares approach to finding the model parameters represent a specific case of *maximum likelihood*, and the over-fitting problem can be understood as a general property of maximum likelihood. By adopting a *Bayesian* approach, the over-fitting problem can be avoided. In a Bayesian model, the *effective* number of parameters adapts automatically to the size of the data set.

In the context of using complex and flexible models with data sets of limited size, one technique that is often used to control over-fitting is *regularisation*, which involves adding a penalty term to the error function to discourage the coefficients from reaching large values. The simplest penalty term takes the form of a sum of squares of all the coefficients, leading to a modified error function of the form
$$
\tilde{E}(\mathbf{w})=\frac{1}{2}\sum^{N}_{n=1}\lbrace y(x_n, \mathbf{w}) - t_n \rbrace^2 + \frac{\lambda}{2}\Vert\mathbf{w}\Vert^2
$$
where the coefficient $\lambda$ governs the relative importance of the regularization term compared with the sum-of-squares error term. The error function can be minimised in closed form. Techniques such as this are known as *shrinkage* methods because they reduce the value of the coefficients. The particular case of a quadratic regularizer is called *ridge regression*. In the context of neural networks, this approach is known as *weight decay*.

We use the `scikit-learn` implementation of ridge regression with polynomial fitting for small data sets.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline

degree = 9
lambdas = [1, np.e**-18] # set value for lamda

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

for i, ax in enumerate(axs):
    
    poly = make_pipeline(PolynomialFeatures(degree=degree), Ridge(alpha=lambdas[i]))
    poly.fit(x.reshape(-1, 1), y)
    y_pred = poly.predict(x_pred.reshape(-1, 1))
    
    ax.scatter(x, y)
    ax.plot(x_pred, y_pred, 'r')
    ax.set_ylim(-1.5, 1.5)
    ax.set_xlabel('$x$', fontsize=12)
    ax.set_ylabel('$y(x)$', fontsize=12)
    ax.set_title('$\ln\lambda = {}, M=9$'.format(int(np.log(lambdas[i]))))
```

For a value of $\ln\lambda=-18$, the over-fitting has been suppressed and we now obtain a much closer representation of the underlying function $\sin(2\pi x)$. If we use too large a value for $\lambda$ then we again obtain a poor fit. Regularization has the desired effect of reducing the magnitude of the coefficients. The choice of the value for $\lambda$ controls the effective complexity of the model and hence determines the degree of over-fitting. The approach of minimising an error function requires a way to determine a suitable value for model complexity. A simple way of achieving this is to take the available data and partition in into a training set to determine $\mathbf{w}$, and a separate *validation* set used to optimise the model complexity (either $M$ or $\lambda$). In many cases, this will prove too wasteful of valuable training data, so we have to seek more sophisticated approaches.
